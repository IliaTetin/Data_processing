## Example: Compound Interest

In economics, the formula for compound interest is often represented as:
$$ A = P\left(1+\frac{r}{n}\right)^{nt}$$

$A$ is the money accumulated after n years, including interest.
$P$ is the principal amount.
$r$ is the annual interest rate (in decimal form).
$n$ is the number of times that interest is compounded per year.
$t$ is the time the money is invested for, in years.

In Python, the integer division operator (//) performs division and truncates the result to the nearest integer towards zero. Unlike the standard division operator (/), it discards the decimal part of the quotient.

When dealing with negative numbers, it's important to note that the result of integer division is rounded towards the more negative number. In other words, it rounds "down," even when the number is negative.

The modulo operator (%) returns the remainder of the division between two integers. This operator is especially useful for a variety of programming and mathematical tasks.  When $n<m$, the result of `n%m` is $n$ itself

It is easy to understand by what algorithm you can find each digit $n$ -significant number $num$: 


`Last digit: (num % 10 ** 1) // 10 ** 0`;

## Working with Floating-Point Numbers in Python

### Floating point numbers
In addition to integers, Python also supports arithmetic operations with fractional or real numbers. For instance, numbers like $\frac{2}{3}, \sqrt{2}, and \pi$ are real numbers and cannot be adequately represented using the integer data type *int*

In computer science, these fractional or real numbers are referred to as **floating-point numbers**. Python employs the data type *float* to handle these numbers.

To convert a string to a floating point number, we use the command `float():`
`num = float(input())`

The following code converts the string 1.2345 to a floating point number: 
`n = float('1.2345')`


````python
n = float('1.2345')
print(n)
````

### Converting between int and float 

Implicit conversion. Any integer (type int) can be used where a floating point number is expected (type float) because Python will automatically convert integers to floats when necessary.

Explicit conversion. A floating point number cannot be implicitly converted to an integer. For such a conversion, you must use an explicit conversion using the command int(). 

````python
num1 = 17.89
num2 = -13.56
num3 = int(num1)
num4 = int(num2)
print(num3)
print(num4)
````

Please note that the conversion of floating point numbers to an integer is done with rounding towards zero, that is

`int(1.7) = 1, int(-1.7) = -1. `

Do not confuse the conversion and rounding operations. Additional commands are used to round floating point numbers.
